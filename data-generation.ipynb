```python
system_instruction = """
Generate a sample entry for fine-tuning a Founder-Investor Matching AI Model.  
Include:  
- A founder profile: Industry, stage, funding required, business model, traction.  
- An investor profile: Preferred industry, stage, investment range, business model, location.  
- A match score (0-100) indicating compatibility.  
- An insights field explaining the reasons behind the match score.  
- Insight must be clear.

Format the output as this format:  
- Score:
- Insight: 
"""
```

```python
def user_instruction(investor_Industry_focus, investor_funding_stage, investor_investment_range, investor_business_model , investor_geographical_preference,founder_Industry, founder_stage, founder_funding_required, founder_business_model, founder_traction):
    
    user_instruction_1 = f"""
     Match the following founder with the investor and determine the best pairings based on compatibility.  

        Founders:
         Founder 1: {founder_Industry}, {founder_stage}, {founder_funding_required}, {founder_business_model}, {founder_traction}.  


        Investors:
         Investor E: {investor_Industry_focus}, {investor_funding_stage}, {investor_investment_range}, {investor_business_model}, {investor_geographical_preference}.  

        Output Requirements:
        Best Pairings: Match each founder with the most compatible investor.  
        Match Score: Display a **compatibility score (0-100)** for each pairing.  
        Insights: Explain why the founder and investor are a strong match or why they aren’t.  
        Output Format: 
        "best_pairs": List of optimal founder-investor matches.  
        "score": Compatibility score for each match.  
        "insights": Explanation of the compatibility.  

        Give the response in the normal text format. First give the Score: and then Insight: . And don't write anything else other than Score and Insight.
     """
    return user_instruction_1
    
```

```python
investor_Industry_focus = "Technology, AI"
investor_funding_stage = "Series A, Series B"
investor_investment_range = "$1M-$10M"
investor_business_model = "B2B, B2C"
investor_geographical_preference = "North America, Europe"

founder_Industry = "AI, SaaS"
founder_stage = "Seed"
founder_funding_required = "$500K"
founder_business_model = "B2B"
founder_traction = "10,000 users, $100K ARR"

ins = user_instruction(
    investor_Industry_focus, 
    investor_funding_stage, 
    investor_investment_range, 
    investor_business_model, 
    investor_geographical_preference, 
    founder_Industry, 
    founder_stage, 
    founder_funding_required, 
    founder_business_model, 
    founder_traction
)
```

```python
import google.generativeai as genai

def get_gemini_response(api_key, user_instruction, system_instruction, model_name="gemini-2.0-flash-lite"):
    GEMINI_API_KEYS = []
    genai.configure(api_key=api_key)
    model = genai.GenerativeModel(model_name)
    response = model.generate_content(f"System: {system_instruction}\nUser: {user_instruction}")
    cleaned_response = re.sub(r'[\*]+', '', response.text)
    return cleaned_response

response = get_gemini_response(random.choice(GEMINI_API_KEYS), ins, system_instruction)
print("Gemini Response:", response)
```

```python
for i in range(NUMBER_OF_DATASET):
    founder_Industry = random.choice(data['Startup_Properties']['Industry'])
    founder_stage = random.choice(data['Startup_Properties']['Stage'])
    founder_funding_required = random.choice(data['Startup_Properties']['Funding_Required'])
    founder_business_model = random.choice(data['Startup_Properties']['Business_Model'])
    founder_traction = random.choice(data['Startup_Properties']['Traction'])

    investor_Industry_focus = random.choice(data['Investor_Properties']['Industry'])
    investor_funding_stage = random.choice(data['Investor_Properties']['Funding_Stage'])
    investor_investment_range = random.choice(data['Investor_Properties']['Investment_Range'])
    investor_business_model = random.choice(data['Investor_Properties']['Business_Model'])
    investor_geographical_preference = random.choice(data['Investor_Properties']['Geographical_Preference'])

    full_details = (
        f"Startup Details: Industry: {founder_Industry}, "
        f"Stage: {founder_stage}, "
        f"Funding Required: {founder_funding_required}, "
        f"Business Model: {founder_business_model}, "
        f"Traction: {founder_traction} | "
        f"Investor Details: Industry: {investor_Industry_focus}, "
        f"Funding Stage: {investor_funding_stage}, "
        f"Investment Range: {investor_investment_range}, "
        f"Business Model: {investor_business_model}, "
        f"Geographical Preference: {investor_geographical_preference}"
    )
    
    print(f"Completed {i}!!!")

    while True:
        api_key = random.choice(GEMINI_API_KEYS)
        try:
            response = get_gemini_response(
                api_key, 
                user_instruction(investor_Industry_focus, investor_funding_stage, investor_investment_range, investor_business_model, investor_geographical_preference, founder_Industry, founder_stage, founder_funding_required, founder_business_model, founder_traction), 
                system_instruction
            ).replace("\n", " ")
            break
        except Exception as e:
            if "exhausted" in str(e).lower():
                print(f"API key {api_key} exhausted, retrying with another key...")
                continue
            else:
                raise e
    
    dataset.append({"Details": full_details, "Response": response})

df = pd.DataFrame(dataset)
df
```
